Load Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
iris
```

Shuffling dataset

```python
import random
indices = list(range(len(iris['data'])))
random.shuffle(indices)

iris['data'] = iris['data'][indices]
iris['target'] = iris['target'][indices]
iris
```

Converting target to is setosa or not

```python
iris['target'] = (iris['target'] == 0).astype(int)
iris
```

Making data split for train and test

```python
split_size = 0.8
split_index = int(np.floor(len(iris['data']) * split_size))
train = {}
train['data'] = iris['data'][:split_index]
train['target'] = iris['target'][:split_index]

test = {}
test['data'] = iris['data'][split_index:]
test['target'] = iris['target'][split_index:]
```

Defining generic Layer class

```python
import numpy as np

class Layer():
  def __init__(self, n_in: int, n_out: int):
    self.weights = np.random.randn(n_in, n_out) * 0.01
    self.bias = np.zeros(n_out)

  def forward(self, samples):
    self.forward_input = samples
    weighted_sum = samples.dot(self.weights) + self.bias
    # print(f'weighted_sum: {weighted_sum}')
    return self.activation(weighted_sum)

  def backward(self, grad):
    # print(f'grad: {grad}')
    # print(f'self.activation_input: {self.activation_input}')
    # print(f'self.derivative(self.activation_input): {self.derivative(self.activation_input)}')
    grad_result = grad * self.derivative(self.activation_input)
    # print(f'grad_result: {grad_result}')

    self.grad_weights = (grad_result * self.forward_input).T
    self.grad_bias = grad_result


    return grad_result

  def optimize(self, learning_rate):
    # print(f'learning_rate: {learning_rate}')
    # print(f'self.grad_weights: {self.grad_weights}')
    # print(f'self.grad_weights * learning_rate: {self.grad_weights * learning_rate}')
    # print(f'self.weights: {self.weights}')
    self.weights -= (self.grad_weights * learning_rate).reshape(self.weights.shape)
    self.grad_weights = 0

    self.bias -= (self.grad_bias * learning_rate).reshape(self.bias.shape)
    self.grad_bias = 0

  def activation(self, value):
    self.activation_input = value
    return np.array(list(map(lambda x: x if x >= 0 else 0.05*x, value)))

  def derivative(self, value):
    return np.array(list(map(lambda x: 1 if x >= 0 else 0.05, value)))

```

Train

```python
layer = Layer(4, 1)

learning_rate = 0.1
xs = range(100)
accuracies = []
for epoch in xs:
  print(f'Epoch {epoch}')
  hits = 0
  for i in range(len(train['data'])):
    train_sample = train['data'][i].reshape(1, 4)
    target_sample = train['target'][i]


    prediction = layer.forward(train_sample)
    prediction = prediction >= 0.5
    error = (prediction - target_sample)**2
    # de / dprediction = 2(prediction - target_sample)
    # print('error:', error)
    # print('train_sample:', train_sample)

    # weight_update = learning_rate * error * train_sample

    # print('layer.weights: ', layer.weights)
    # print('layer.weights.shape: ', layer.weights.shape)
    # print('layer.bias: ', layer.bias)
    # print('layer.bias.shape: ', layer.bias.shape)

    layer.backward(2 * (prediction - target_sample))
    layer.optimize(learning_rate)

    # print('weight_update: ', weight_update)
    # print('weight_update.shape: ', weight_update.shape)

    # layer.weights += weight_update.T
    # layer.bias += learning_rate * error.reshape(-1)
    hits += prediction == target_sample

  accuracies.append((hits / len(train['data'])).reshape(-1)[0])

  print(f'Weights: \n{layer.weights}\nBias: {layer.bias}')
  print(f'Accuracy: {hits / len(train["data"])}')
plt.plot(xs, accuracies)
```

Test

```python
hits = 0
for i in range(len(test['data'])):
  test_sample = test['data'][i].reshape(1, 4)
  target_sample = test['target'][i]

  prediction = layer.forward(test_sample)
  prediction = prediction >= 0.5

  hits += prediction == target_sample
print('Accuracy: ', hits/len(test['data']))
```
